```python
#1- Coffee Maker Program 
resources = {
    "water": 300,
    "milk": 200,
    "coffee": 100,
    "money": 0.0
}

prices = {
    "espresso": 1.5,
    "latte": 2.5,
    "cappuccino": 3.0
}

ingredients = {
    "espresso": {"water": 50, "coffee": 18},
    "latte": {"water": 200, "milk": 150, "coffee": 24},
    "cappuccino": {"water": 250, "milk": 100, "coffee": 24}
}

while True:
    user_input = input("What would you like? (espresso/latte/cappuccino): ")
    if user_input == "off":
        print("Turning off the coffee machine. Goodbye!")
        break

    elif user_input == "report":
        print("Water: ", resources['water'], "ml")
        print("Milk: ", resources['milk'], "ml")
        print("Coffee: ", resources['coffee'], "g")
        print("Money: $", resources['money'])

    elif user_input in prices:
        enough_resources = True
        for item in ingredients[user_input]:
            if resources[item] < ingredients[user_input][item]:
                print("Sorry there is not enough", item)
                enough_resources = False
        
        if enough_resources:
            print("The cost of", user_input, "is $ ",prices[user_input], "Please insert coins.")
            quarters = int(input("How many quarters? ")) * 0.25
            dimes = int(input("How many dimes? ")) * 0.10
            nickels = int(input("How many nickels? ")) * 0.05
            pennies = int(input("How many pennies? ")) * 0.01
            total_inserted = quarters + dimes + nickels + pennies
            print("You inserted: $", int(total_inserted))

            if total_inserted < prices[user_input]:
                print("Sorry that's not enough money. Money refunded.")
            else:
                change = total_inserted - prices[user_input]
                if change > 0:
                    print("Here is $", int(change), "dollars in change.")
        
                for item in ingredients[user_input]:
                    resources[item] -= ingredients[user_input][item]

                resources["money"] += prices[user_input]

                print("Here is your", user_input, "Enjoy!")
    
    else:
        print("Invalid input. Please choose espresso, latte, cappuccino, or type 'off' to turn off the machine.")
```

```python
# ATM System
balance = 100000 

while True:
    print("\nWelcome to the ATM!")
    print("Please choose an option:")
    print("1. Display current balance")
    print("2. Deposit money")
    print("3. Withdraw money")
    print("4. Exit")

    choice = input("Enter your choice (1-4): ")
    
    if choice == '1':
        print("Your current balance is: $", balance)

    elif choice == '2':
        try:
            deposit_amount = float(input("Enter the amount to deposit: "))
            if deposit_amount > 0:
                balance += deposit_amount
                print("You have deposited: $", deposit_amount)
                print("Your updated balance is: $", balance)
            else:
                print("Please enter a positive amount.")
        except ValueError:
            print("Invalid input. Please enter a numeric value.")

    elif choice == '3':
        try:
            withdraw_amount = float(input("Enter the amount to withdraw: "))
            if withdraw_amount > 0:
                if withdraw_amount <= balance:
                    balance -= withdraw_amount
                    print("You have withdrawn: $", withdraw_amount)
                    print("Your updated balance is: $", balance)
                else:
                    print("Insufficient funds.")
            else:
                print("Please enter a positive amount.")
        except ValueError:
            print("Invalid input. Please enter a numeric value.")

    elif choice == '4':
        print("Thank you for using the ATM. Goodbye!")
        break

    else:
        print("Invalid choice. Please select a valid option (1-4).")
```
