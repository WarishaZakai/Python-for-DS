```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('googleplaystore.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#ranges given = median
#ranges not given = mean
```

```python
df['Price'].unique()
```

```python
#Replace 'everyone' with 0 and then change the data type of column to float
df['Price'] = df['Price'].replace('Everyone','0')
```

```python
df['Price'].unique()
```

```python
df.Price.dtype
```

```python
df['Price'] = df['Price'].str.replace('$','').astype(float)
```

```python
df['Price'].unique()
```

```python
df.Price.dtype
```

```python
df['Size'].unique()
```

```python
df['Size'] = df['Size'].replace('Varies with device','0')
```

```python
df['Size'].unique()
```

```python
df['Size'] = df['Size'].str.replace('M', '').str.replace('k','').str.replace('+','').str.replace(',','').astype(float)
```

```python
df['Size'].unique()
```

```python
df.Size.dtype
```

```python
df.shape
```

```python
df = df.drop_duplicates(subset=['App'])
```

```python
df.shape
```

```python
df
```

```python
df['Rating'].unique()
```

```python
df['Rating'].mean()
```

```python
df['Rating'].median()
```

```python
#difference of mean and median is larger then fill with median
```

```python
df['Rating'] = df['Rating'].fillna(df['Rating'].mean())
```

```python
df['Rating'].unique()
```

```python
df['Reviews'].unique()
```

```python
df['Reviews'] = df['Reviews'].str.replace('M','').astype(float)
```

```python
df['Reviews'].dtype
```

```python
#data normalization
#we always normalize data in between 0 and 1
#(current value - min value)/(max value - min value)
```

```python
df['Reviews'] = (df['Reviews'] - df['Reviews'].min()) / (df['Reviews'].max() - df['Reviews'].min())
```

```python
df.head()
```

```python
#popularity = Rating * Reviews
df['Popularity'] = df['Rating'] * df['Reviews']
```
